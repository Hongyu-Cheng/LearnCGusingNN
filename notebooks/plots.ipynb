```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu_grid = np.linspace(0,1,101)
mu_treesize_raw = np.load('../data/mu_treesize_CG_and_GMI.npy')
mu_treesize = []
for i in range(len(mu_grid)):
    for j in range(100):
        mu_treesize.append((mu_treesize_raw[i]))
mu_grid = np.linspace(0,1,len(mu_treesize))
```

```python
plt.figure(figsize=(6, 6))

plt.plot(mu_grid, mu_treesize, label='Max weighted auxiliary scores', color='navy', linewidth=2)

plt.axhline(y=176.82, color='mediumseagreen', linestyle='-.', linewidth=2, label='ReLU neural network')
plt.axhline(y=191.871, color='mediumpurple', linestyle='-.', linewidth=2, label='Linear threshold neural network')


plt.legend(loc='upper left', frameon=False)

plt.grid(True, linestyle='--', alpha=0.5)

plt.xlabel('$\mu$', fontsize=14)
plt.ylabel('Averaged treesize over 1000 instances', fontsize=14)
plt.title(' ', fontsize=16)

plt.xticks(np.arange(0, 1.1, 0.1))
plt.yticks(np.arange(200, 641, 40))

plt.tight_layout() 

plt.savefig('../figures/mu_treesize_with_comparisons.pdf', format='pdf', dpi=1000)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

moving_return_ReLU = np.load('../data/moving_returns_ReLU.npy')
moving_return_LTNN = np.load('../data/moving_returns_LTNN.npy')

sns.set(style='whitegrid')

plt.figure(figsize=(10, 6), dpi=300)

plt.plot(moving_return_ReLU, label='ReLU', color='blue')
plt.plot(moving_return_LTNN, label='LTNN', color='green')

plt.xlabel('Episodes', fontsize=14)
plt.ylabel('Moving average of rewards', fontsize=14)

plt.legend(fontsize=12)

plt.grid(True)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.savefig('../figures/moving_returns_comparison.png', bbox_inches='tight', dpi=300)
plt.show()
```
